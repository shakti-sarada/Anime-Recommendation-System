```python
import pandas as pd
import ast

# Assuming your data is in the 'anime_data' DataFrame and the relevant columns are in list form

# Function to extract 'name' field from the dictionary format in each column
def extract_names(column_data):
    # Safely evaluate the string to a Python object (list of dictionaries)
    if pd.isna(column_data):
        return []  # Return an empty list if the value is NaN
    try:
        data_list = ast.literal_eval(column_data)
        # Extract the 'name' field from each dictionary
        return [item['name'] for item in data_list]
    except (ValueError, SyntaxError):
        return []  # Handle any parsing issues by returning an empty list

# Columns to apply the transformation on
columns_to_transform = ['producers', 'licensors', 'studios', 'genres', 'themes']

# Apply the transformation to each of the specified columns
for column in columns_to_transform:
    anime_data[column] = anime_data[column].apply(extract_names)
```

```python
def remove_space(word):
    # Replace spaces in the entire string
    return word.replace(" ", "")
```

```python
# Function to convert list of names into a comma-separated string
def list_to_string(name_list):
    return ', '.join(name_list)

# Apply the transformation to the specified columns
for column in columns_to_transform:
    anime_data[column] = anime_data[column].apply(list_to_string)
```

```python
anime_data.head()
```

```python
# List of columns to apply the remove_space function
columns_to_modify = ['producers','licensors','studios','genres','themes']

# Apply remove_space to each specified column
for col in columns_to_modify:
    anime_data[col] = anime_data[col].apply(remove_space)
```

```python
anime_data.head()
```

```python
# Convert list-like objects to strings and ensure all columns are strings
def convert_to_string(x):
    if isinstance(x, list):
        return ', '.join(x)
    return str(x)

# Apply the conversion function to each specified column
for col in ['score', 'producers','licensors', 'studios', 'licensors', 'themes']:
    anime_data[col] = anime_data[col].apply(convert_to_string)

# Concatenate the specified columns into 'tags'
anime_data['tags'] = anime_data[['score', 'producers','licensors', 'studios', 'licensors', 'themes']].agg(' '.join, axis=1)

# Convert the 'tags' column from string to list
anime_data['tags'] = anime_data['tags'].apply(lambda x: x.split())
```

```python
# Function to clean up extra commas and spaces in list elements
def clean_tag(tag):
    return tag.strip(', ')  # Remove leading/trailing commas and spaces

# Apply the cleaning function to each element in the 'tags' column
anime_data['tags'] = anime_data['tags'].apply(lambda lst: [clean_tag(tag) for tag in lst])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# definning fuction for lemmetization
def lems(text):
    l = []
    for i in text.split():
        l.append(lemmatizer.lemmatize(i))
        
    return " ".join(l)
```

```python
# applying lemmetiztion on the tags column
new_df['tags'] = new_df['tags'].apply(lems)
```

```python
#defining max features

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=15000,stop_words='english')
```

```python
# converting the data to vectors

vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarly = cosine_similarity(vector)
```

```python
similarly
```

```python
similarly.shape
```

```python
# Define the search value
search_value = 'Naruto'  # Replace with the desired search term

# Filter the DataFrame based on the search value
matching_rows = new_df[new_df['title'] == search_value]

# Check if there are any matching rows
if not matching_rows.empty:
    # Always access the first match
    index_value = matching_rows.index[0]
    print(f"Index: {index_value}, Search Name: '{search_value}'")
else:
    print(f"No rows found with 'Name' equal to '{search_value}'")
```

```python
def recommend(anime):
    index = new_df[new_df['title'] == anime].index[0]
    distances = sorted(list(enumerate(similarly[index])),reverse=True, key=lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Naruto')
```
